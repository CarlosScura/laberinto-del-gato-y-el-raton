```python
# importamos las librerias necesarias
import random

# creamos las variables que necesitamos
laberinto = []
tamano = 5
pos_cat = (0,0)
pos_rat = (0,4)

for i in range(tamano):
    fila=[]
    for j in range(tamano):
        fila.append(' ')
        laberinto.append(fila)

def mostrar_lab():
    global laberinto
    for i in range(tamano):
        print(laberinto[i])

mostrar_lab()
```

```python
# rellenamos el laberinto con paredes y espacios vacios
for i in range(tamano):
    # creamos las filas del laberinto
    fila=[]
    # rellenamos los espacios.
    for j in range(tamano):
        if random.random() < 0.3:
            # Agrega una pared
            fila.append('|')
        else:
            # agrega un espacio vacio
            fila.append(' ')
    laberinto.append(fila)
for i in range(tamano):
    print(laberinto[i])
```

```python
# con el laberinto creado agregamos la salida del laberinto

# buscamos los espacios libres en la ultima fila del laberinto
libres = []
for l in range(tamano):
    if laberinto[tamano-1][l] != '|':
        libres.append(l)

# si hay espacios libres, elegimos uno al azar y colocamos la salida 'S'
if libres:
    l = random.choice(libres)
    laberinto[tamano-1][l] = 'S'

# imprimimos el laberinto

for i in range(tamano):
    print(laberinto[i])
        
```

# minimax

```python
import math

def minimax(estado, depth, maximizing_player):
    """
    estado: el estado actual del juego (posiciones gato/ratón, tablero)
    depth: cuántos turnos futuros simulamos
    maximizing_player: True si es el turno del ratón (quiere escapar),
                       False si es el turno del gato (quiere atrapar)
    """
    # 1. Condición de parada
    if depth == 0 or is_terminal(state):
        return evaluate(state), None

    best_move = None

    # 2. Turno del ratón (maximiza)
    if maximizing_player:
        max_eval = -math.inf
        for move in legal_moves(state, player="mouse"):
            child_state = apply_move(state, move, player="mouse")
            eval_score, _ = minimax(child_state, depth-1, False)
            if eval_score > max_eval:
                max_eval = eval_score
                best_move = move
        return max_eval, best_move

    # 3. Turno del gato (minimiza)
    else:
        min_eval = math.inf
        for move in legal_moves(state, player="cat"):
            child_state = apply_move(state, move, player="cat")
            eval_score, _ = minimax(child_state, depth-1, True)
            if eval_score < min_eval:
                min_eval = eval_score
                best_move = move
        return min_eval, best_move
```
