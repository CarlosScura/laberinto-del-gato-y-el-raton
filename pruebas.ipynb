## Prueba de creacion de laberinto.

Creamos la grilla del laberinto.
Agregamos las paredes del laberinto.
Agregamos la entrada y la salida del laberinto.

```python
# importamos las librerias necesarias
import random

# creamos las variables que necesitamos
laberinto = []
tamano = 5
pos_cat = (0,0)
pos_rat = (0,4)

for i in range(tamano):
    fila=[]
    for j in range(tamano):
        fila.append(' ')
        laberinto.append(fila)

def mostrar_lab():
    global laberinto
    for i in range(tamano):
        print(laberinto[i])

mostrar_lab()
```

```python
# rellenamos el laberinto con paredes y espacios vacios
for i in range(tamano):
    # creamos las filas del laberinto
    fila=[]
    # rellenamos los espacios.
    for j in range(tamano):
        if random.random() < 0.3:
            # Agrega una pared
            fila.append('|')
        else:
            # agrega un espacio vacio
            fila.append(' ')
    laberinto.append(fila)
for i in range(tamano):
    print(laberinto[i])
```

```python
# con el laberinto creado agregamos la salida del laberinto

# buscamos los espacios libres en la ultima fila del laberinto
libres = []
for l in range(tamano):
    if laberinto[tamano-1][l] != '|':
        libres.append(l)

# si hay espacios libres, elegimos uno al azar y colocamos la salida 'S'
if libres:
    l = random.choice(libres)
    laberinto[tamano-1][l] = 'S'

# imprimimos el laberinto

for i in range(tamano):
    print(laberinto[i])
        
```

```python
# agregamos al gato y al ratón.

libres = []
for f in range(tamano):
    for c in range(tamano):
        if laberinto[f][c] == ' ':
            libres.append((f,c)) 

# si hay espacios libres, elegimos uno al azar y colocamos al gato 'G'
if libres:
    l = random.choice(libres)
    laberinto[l[0]][l[1]] = 'G'
    pos_cat = l
    libres.remove(l)

# si hayh espacios libres, elegimos uno al azar y colocamos al ratón 'R'
if libres:
    l = random.choice(libres)
    laberinto[l[0]][l[1]] = 'R'
    pos_rat = l
    libres.remove(l)

for i in range(tamano):
    print(laberinto[i])
```

# minimax

```python
import math

# definimos el tamaño del tablero y el estado inicial del juego
BOARD_SIZE = (5, 5)

state = {
    "cat": (0, 0),      # esquina superior izquierda
    "mouse": (0, 1),    # esquina superior derecha
    "board_size": BOARD_SIZE
}

def is_terminal(state):
    # termina si gato atrapa al ratón
    return state["cat"] == state["mouse"]

def evaluate(state):
    # distancia Manhattan entre gato y ratón
    cat = state["cat"]
    mouse = state["mouse"]
    dist = abs(cat[0] - mouse[0]) + abs(cat[1] - mouse[1])
    return dist  # cuanto más lejos, mejor para el ratón

def legal_moves(state, player):
    x, y = state[player]
    moves = []
    for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:  # 4 direcciones
        nx, ny = x+dx, y+dy
        if 0 <= nx < state["board_size"][0] and 0 <= ny < state["board_size"][1]:
            moves.append((nx, ny))
    return moves

def apply_move(state, move, player):
    new_state = state.copy()
    new_state[player] = move
    return new_state

# implementamos el algoritmo minimax para que el gato y el raton puedan tomar decisiones inteligentes

def minimax(state, depth, maximizing_player):
    """
    state: el estado actual del juego (posiciones gato/ratón, tablero)
    depth: cuántos turnos futuros simulamos
    maximizing_player: True si es el turno del ratón (quiere escapar),
                       False si es el turno del gato (quiere atrapar)
    """
    # 1. Condición de parada
    if depth == 0 or is_terminal(state):
        return evaluate(state), None

    best_move = None

    # 2. Turno del ratón (maximiza)
    if maximizing_player:
        max_eval = -math.inf
        for move in legal_moves(state, player="mouse"):
            child_state = apply_move(state, move, player="mouse")
            eval_score, _ = minimax(child_state, depth-1, False)
            if eval_score > max_eval:
                max_eval = eval_score
                best_move = move
        return max_eval, best_move

    # 3. Turno del gato (minimiza)
    else:
        min_eval = math.inf
        for move in legal_moves(state, player="cat"):
            child_state = apply_move(state, move, player="cat")
            eval_score, _ = minimax(child_state, depth-1, True)
            if eval_score < min_eval:
                min_eval = eval_score
                best_move = move
        return min_eval, best_move

# ejemplo de uso: el ratón decide su mejor movimiento

score, best_move = minimax(state, depth=2, maximizing_player=True)
print("Mejor movimiento del ratón:", best_move, "con score:", score)
score, best_move = minimax(state, depth=2, maximizing_player=False)
print("Mejor movimiento del gato:", best_move, "con score:", score)
```
